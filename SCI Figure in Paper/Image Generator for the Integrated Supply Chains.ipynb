```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("Figure_1,2.csv")


# Define marker styles and colors for each type and supply chain
marker_styles = {
    'Supplier': 's',  # square
    'Customer': 'P',  # star
    'Distribution Centre': '^',  # triangle
    'Branch': 'o'  # circle
}

color_map = {
    1: 'blue',  # Supply Chain 1
    2: 'gold',    # Supply Chain 2
}

# Create the plot
plt.figure(figsize=(8, 5))

for vertex_type in data['Type'].unique():
    for supply_chain in data['Supply_Chain_Numer'].unique():
        subset = data[(data['Type'] == vertex_type) & (data['Supply_Chain_Numer'] == supply_chain)]
        plt.scatter(
            subset['Longitude'], subset['Latitude'],
            marker=marker_styles[vertex_type],
            color=color_map[supply_chain],
            label=f'{vertex_type} - Supply Chain {supply_chain}',
            s=100  # Size of the marker
        )

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Supply Chain Network Visualization')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
# data = pd.read_csv("Figure_1,2.csv")
data = pd.read_csv("Figure_4,5.csv")

# Define marker styles and colors for each type and supply chain
marker_styles = {
    'Supplier': 's',  # square
    'Customer': 'P',  # star
    'Distribution Centre': '^',  # triangle
    'Branch': 'o'  # circle
}

color_map = {
    1: 'blue',  # Supply Chain 1
    2: 'gold',    # Supply Chain 2
    4: 'red',  # Supply Chain 4
    5: 'orange',    # Supply Chain 5
}

# Create the plot with extended width to accommodate the legend
plt.figure(figsize=(8, 3.5))  # Increase the width to 12 to add space on the right

# Set font family to Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'

for vertex_type in data['Type'].unique():
    for supply_chain in data['Supply_Chain_Numer'].unique():
        subset = data[(data['Type'] == vertex_type) & (data['Supply_Chain_Numer'] == supply_chain)]
        plt.scatter(
            subset['Longitude'], subset['Latitude'],
            marker=marker_styles[vertex_type],
            color=color_map[supply_chain],
            label=f'{vertex_type} - Supply Chain {supply_chain}',
            s=100  # Size of the marker
        )

plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.title('Supply Chain Network Visualization', fontsize=14)

# Position the legend outside the plot area
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)
plt.grid(True)
# plt.show()
plt.savefig('supply_chain_network.png', dpi=300, bbox_inches='tight')
```
